##**Pracitcal 8**

**AIM:**  Write a program to do set operations from data of your dataset: (i)intersection (ii)union (iii)difference (iv)symmetric difference (v)check s1 is a subset of s2(vi)check if s1 is a superset of s2(vii)find whether two sets are disjoint or not(viii)find all subsets of a set without using itertools


```python
a={"uber","retweet","thumbnail","photos","username"}
hashtags={"uber","aeroports","uberflies","ubergate"}
print("Original Sets:")
print("A:",a)
print("B:",hashtags)

print("-----------------------------------------------------------------------")
print("INTERSECTION")
def intersection():
  ch=a.intersection(hashtags)
  return ch
print(intersection())

print("-----------------------------------------------------------------------")
print("UNION")
def union():
  ch=a.union(hashtags)
  return ch
print(union())

print("-----------------------------------------------------------------------")
print("DIFFERENCE")
def difference():
  ch=a.difference(hashtags)
  return ch
print(difference())

print("-----------------------------------------------------------------------")
print("SYMMETRIC DIFFERENCE")
def symmetric_differences():
  ch=a.symmetric_difference(hashtags)
  return ch
print(symmetric_differences())

print("-----------------------------------------------------------------------")
print("SUBSET")
def subset():
  ch=a.issubset(hashtags)
  return ch
print(subset())

print("-----------------------------------------------------------------------")
print("SUPERSET")
def superset():
  ch=a.issuperset(hashtags)
  return ch
print(superset())

print("-----------------------------------------------------------------------")
print("DISJOINT")
def disjoint():
  ch=a.isdisjoint(hashtags)
  return ch 
print(disjoint())

print("-----------------------------------------------------------------------")
print("POWER SET")
def subsets(a):
  if a == []:
    return [[]]
  else:
    b = subsets(a[1:])
    return b + [[a[0]] + c for c in b]
ai=list(a)
thesubsets = subsets(ai)
print(thesubsets)



#OUTPUT
```

## <b>*Practical* 16</b>
Write a program to test data using chi-square.


```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
 
# contingency table
table = [tweets['Avg_retweeted_tweet'], tweets['Least_retweeted_tweet']]
 
stat, p, dof, expected = chi2_contingency(table)
print('dof= %d' % dof)
print('stat: %.3f' % stat)
 
# interpret test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
if abs(stat) >= critical:
 print('Dependent (reject H0)')
else:
 print('Independent (fail to reject H0)')

 #OUTPUT
```

```python
#TF calculation
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
```

```python
tfA = computeTF(numOfWordsA, bagOfWordsA)
tfB = computeTF(numOfWordsB, bagOfWordsB)
```

```python
def computeIDF(documents):
    import math
    N = len(documents)
    
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict
```

```python
idfs = computeIDF([numOfWordsA, numOfWordsB])
```

```python
#TF-IDF
def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf
```
